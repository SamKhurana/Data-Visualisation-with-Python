```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ibm.csv")
df.head()
```

```python
df.drop(["Type","Coverage","AREA","REG","DEV"],axis =1, inplace = True)
df.head()
```

```python
df.rename(columns = {"OdName":"Country","AreaName":"Continent","RegName":"Region"}, inplace = True)
df.head()
```

```python
df.set_index(["Country"], inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.tail()
```

```python
df.drop(["Unknown","Total"], axis = 0, inplace = True)
df.tail()
```

```python
df.columns = list(map(str,df.columns))
```

```python
years = list(map(str,range(1980,2014)))
years
```

```python
df["Total"] = df.sum(axis = 1)
df.head()
```

```python
df.sort_values(by = "Total", ascending = False, inplace = True)
df.head()
```

```python
df.top5 = df.head(5)
df.top5
```

```python
df.top5 = df.top5[years].transpose()
df.top5
```

```python
df.top5.plot(kind = "line", figsize = (10,8))
plt.legend(bbox_to_anchor = (1.05,1))
plt.title("Immigration Trend of Top 5 Countries")
```

```python
df.top5.plot(kind = "area", stacked = False, figsize = (10,8))
plt.legend(bbox_to_anchor = (1.05,1))
plt.title("Immigration Trend of Top 5 Countries")
```

```python
df.top5.plot(kind = "area", stacked = False,alpha = 0.5, figsize = (10,8))
plt.legend(bbox_to_anchor = (1.05,1))
plt.title("Immigration Trend of Top 5 Countries")
```

```python
df_ci = df.loc[["China","India"], years]
df_ci
```

```python
df_ci = df_ci.transpose()
df_ci
```

```python
df_continent = df.groupby("Continent").sum()
df_continent
```

```python
colors_list =['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink'] 
df_continent["Total"].plot(kind ="pie", shadow = "True", figsize = (8,6), colors = colors_list, labels = None, )
plt.legend(df_continent.index, bbox_to_anchor = (1.05,1), loc = "best")
```
